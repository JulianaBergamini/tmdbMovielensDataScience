```python
import pandas as pd
```

```python
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.original_language.unique()
```

```python
contagem_de_lingua = tmdb['original_language'].value_counts().to_frame().reset_index()
contagem_de_lingua.columns = ['original_language', 'total']
contagem_de_lingua.head()
```

```python
import seaborn as sns
sns.barplot(x = 'original_language', y = 'total', data = contagem_de_lingua)
```

```python
sns.catplot(x = 'original_language', kind = 'count', data = tmdb)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(contagem_de_lingua['total'], labels = contagem_de_lingua['original_language'])
```

```python
total_por_lingua = tmdb['original_language'].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua.loc['en']
total_do_resto = total_geral - total_de_ingles
print (total_de_ingles, total_do_resto)
```

```python
dados = {
    'lingua': ['ingles', 'outros'],
    'total' : [total_de_ingles, total_do_resto]
}
dados
```

```python
pd.DataFrame(dados)
```

```python
sns.barplot(x = 'lingua', y = 'total', data = dados)
```

```python
plt.pie(dados["total"], labels = dados["lingua"])
```

```python
tmdb.query("original_language == 'en'")
```

```python
filmes_sem_lingua_original_em_ingles = tmdb.query("original_language != 'en'").original_language.value_counts()
print(filmes_sem_lingua_original_em_ingles)
```

```python
filmes_sem_ingles = tmdb.query("original_language != 'en'")

sns.catplot(x = "original_language",
            data = filmes_sem_ingles,
            kind='count', 
            aspect=2,
            order = filmes_sem_lingua_original_em_ingles.index,
            palette ='copper')
```

```python
notas = pd.read_csv('ratings.csv')
notas.head()
```

```python
notas.columns = ['usuarioId', 'filmeId', 'nota', 'momento']
notas.head()
```

```python
notas_do_toy_story = notas.query('filmeId==1')
notas_jumanji = notas.query('filmeId==2')
print(len(notas_do_toy_story), len(notas_jumanji))
```

```python
print("A nota média de Toy Story é %.2f" % notas_do_toy_story.nota.mean())
print("A nota média de Jumanji é %.2f" % notas_jumanji.nota.mean())
```

```python
print("A mediana de Toy Story é %.2f" % notas_do_toy_story.nota.median())
print("A mediana de Jumanji é %.2f" % notas_jumanji.nota.median())
```
